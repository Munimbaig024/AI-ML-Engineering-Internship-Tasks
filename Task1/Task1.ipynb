```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Iris dataset
df = sns.load_dataset('iris') 

```

```python

# 1. Dataset Overview
print("Shape of the dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst 5 rows of the dataset:\n", df.head())

```

```python

# 2. Summary and Info
print("\nDataset Info:")
print(df.info())

print("\nDescriptive Statistics:\n")
print(df.describe())
```

## 3. Visualizations

### Scatter Plot: Sepal Length vs Petal Length

```python


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()


```

### Histograms for each numerical feature


```python
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()

```

### Box Plots for each feature by species

```python
 
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1], 1):  
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x='species', y=column)
    plt.title(f'Box plot of {column} by species')
plt.tight_layout()
plt.show()
```

## Iris Dataset EDA: Key Takeaways
###  Overview
The Iris dataset has 150 samples, each with four measurements (sepal length/width, petal length/width) and a species label (setosa, versicolor, or virginica).
### What We Found
- No missing data: Everything is complete and ready for analysis.
- Petal size matters: Petal length and width are the best features for telling the species apartsetosa especially stands out with much smaller petals.
- Sepal measurements: These help a bit, but theres more overlap between species.
- Clear clusters: The scatter plot shows that the three species mostly form their own groups, with setosa being the easiest to spot.
### Final Thoughts
The Iris dataset is clean and shows clear differences between species, especially in petal size. This makes it a great example for learning about data analysis and building simple machine learning models.
